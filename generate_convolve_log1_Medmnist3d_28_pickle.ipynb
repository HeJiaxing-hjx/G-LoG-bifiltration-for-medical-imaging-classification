# Example for `adrenalmnist3d`

```python
"""
import dataset
"""

data_flag = 'adrenalmnist3d'
download = False


info = INFO[data_flag]
task = info['task']
print(task)
n_channels = info['n_channels']
n_classes = len(info['label'])

DataClass = getattr(medmnist, info['python_class'])

data_transform = None

train_dataset = DataClass(split='train', transform=data_transform, size =28, download=download)
test_dataset = DataClass(split='test', transform=data_transform,size=28, download=download)
val_dataset = DataClass(split='val', transform=data_transform,size=28, download=download)


train_list = []
train_label_list = []
test_list = []
test_label_list = []
val_list = []
val_label_list = []
for i in train_dataset:
    train_list.append(i[0])
    train_label_list.append(i[1])
for j in test_dataset:
    test_list.append(j[0])
    test_label_list.append(j[1])
for k in val_dataset:
    val_list.append(k[0])
    val_label_list.append(k[1])
    
train_label_list = np.array(train_label_list)
train_label_list_tran = np.squeeze(train_label_list)
test_label_list = np.array(test_label_list)
test_label_list_tran = np.squeeze(test_label_list)
val_label_list = np.array(val_label_list)
val_label_list_tran = np.squeeze(val_label_list)
```

```python
from scipy import ndimage
import scipy as sp
from multipers.filtrations import Cubical
import pickle


resolution = [28,28,28]

"""
save pickle
"""
pickle_path_withker8_train = os.path.join(pickle_path_withker8,'train')
for i in tqdm(range(len(train_list))):
    """
    or
    """
    #img1 = train_list[i][0]
    """
    s05
    """
    img1 = ndimage.gaussian_filter(train_list[i][0].astype(float), sigma=0.5)   
    """
    s1
    """
    #img1 = ndimage.gaussian_filter(train_list[i][0].astype(float), sigma=1)
    """
    s15
    """
    #img1 = ndimage.gaussian_filter(train_list[i][0].astype(float), sigma=1.5)
    
    img3 = ndimage.gaussian_laplace(train_list[i][0].astype(float), sigma=1)
    bifiltration = np.empty(shape=(*resolution, 2))
    bifiltration[:,:,:,0] = img1
    bifiltration[:,:,:,1]=img3
    s = Cubical(bifiltration)
    mod = mp.module_approximation(s)
    file_path = os.path.join(pickle_path_withker8_train,'train_{}.pickle'.format(i))
    with open(file_path, "wb") as file:
        pickle.dump(mod, file)
        
pickle_path_withker8_test = os.path.join(pickle_path_withker8,'test')
for i in tqdm(range(len(test_list))):
     """
    or
    """
    #img1 = test_list[i][0]
    """
    s05
    """
    img1 = ndimage.gaussian_filter(test_list[i][0].astype(float), sigma=0.5)   
    """
    s1
    """
    #img1 = ndimage.gaussian_filter(test_list[i][0].astype(float), sigma=1)
    """
    s15
    """
    #img1 = ndimage.gaussian_filter(test_list[i][0].astype(float), sigma=1.5)
    
    img3 = ndimage.gaussian_laplace(test_list[i][0].astype(float), sigma=1)
    bifiltration = np.empty(shape=(*resolution, 2))
    bifiltration[:,:,:,0] = img1
    bifiltration[:,:,:,1]=img3
    s = Cubical(bifiltration)
    mod = mp.module_approximation(s)
    file_path = os.path.join(pickle_path_withker8_test,'test_{}.pickle'.format(i))
    with open(file_path, "wb") as file:
        pickle.dump(mod, file)
        
pickle_path_withker8_val = os.path.join(pickle_path_withker8,'val')
for i in tqdm(range(len(val_list))):
    """
    or
    """
    #img1 = val_list[i][0]
    """
    s05
    """
    img1 = ndimage.gaussian_filter(val_list[i][0].astype(float), sigma=0.5)   
    """
    s1
    """
    #img1 = ndimage.gaussian_filter(val_list[i][0].astype(float), sigma=1)
    """
    s15
    """
    #img1 = ndimage.gaussian_filter(val_list[i][0].astype(float), sigma=1.5)

    img3 = ndimage.gaussian_laplace(val_list[i][0].astype(float), sigma=1)
    bifiltration = np.empty(shape=(*resolution, 2))
    bifiltration[:,:,:,0] = img1
    bifiltration[:,:,:,1]=img3
    s = Cubical(bifiltration)
    mod = mp.module_approximation(s)
    file_path = os.path.join(pickle_path_withker8_val,'val_{}.pickle'.format(i))
    with open(file_path, "wb") as file:
        pickle.dump(mod, file)
```
