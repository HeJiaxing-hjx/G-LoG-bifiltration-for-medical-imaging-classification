```python
import multipers as mp
import numpy as np
import pickle
from tqdm import tqdm
```

```python
import os
# 获取当前工作目录路径
current_dir = os.getcwd()

path_pers = os.path.join(current_dir,'pers_s05_log1')
#l2 band = 0.1
# 获取当前文件的绝对路径
parent_path = os.path.abspath('..')

# 获取上一级目录
grandparent_dir = os.path.dirname(parent_path)
pickle_path = os.path.join(grandparent_dir,'pickle_adrenalmnist3d/pickle_withs05_log1')
```

```python



#train
pickle_train_path = os.path.join(pickle_path,'train')
files_train = os.listdir(pickle_train_path)
#files_train.sort(key=lambda x:int(x[9:].split('.')[0]))
files_train.sort(key=lambda x:int(x[6:].split('.')[0]))
#print(len(files_train))
#print(files_train)
train_pers0 = []
train_pers1 = []
train_pers2 = []
for f1 in tqdm(files_train):
    f1_path = os.path.join(pickle_train_path,f1)
    with open(f1_path, "rb") as file:
        loaded_s = pickle.load(file)
        #load_train.append(loaded_s)
    train_pers0.append(loaded_s.representation(degrees=[0], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    train_pers1.append(loaded_s.representation(degrees=[1], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    train_pers2.append(loaded_s.representation(degrees=[2], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
train_pers0 = np.array(train_pers0)
train_pers1 = np.array(train_pers1)
train_pers2 = np.array(train_pers2)

pers0_train_path = os.path.join(path_pers,'train_pers_001/pers_train_h0.npy')
pers1_train_path = os.path.join(path_pers,'train_pers_001/pers_train_h1.npy')
pers2_train_path = os.path.join(path_pers,'train_pers_001/pers_train_h2.npy')
np.save(pers0_train_path,train_pers0)
np.save(pers1_train_path,train_pers1)
np.save(pers2_train_path,train_pers2)

#test
pickle_test_path = os.path.join(pickle_path,'test')
files_test = os.listdir(pickle_test_path)
#files_train.sort(key=lambda x:int(x[9:].split('.')[0]))
files_test.sort(key=lambda x:int(x[5:].split('.')[0]))
# print(len(files_train))
# print(files_train)
test_pers0 = []
test_pers1 = []
test_pers2 = []
for f1 in tqdm(files_test):
    f1_path = os.path.join(pickle_test_path,f1)
    with open(f1_path, "rb") as file:
        loaded_s = pickle.load(file)
        #load_train.append(loaded_s)
    test_pers0.append(loaded_s.representation(degrees=[0], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    test_pers1.append(loaded_s.representation(degrees=[1], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    test_pers2.append(loaded_s.representation(degrees=[2], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
test_pers0 = np.array(test_pers0)
test_pers1 = np.array(test_pers1)
test_pers2 = np.array(test_pers2)

pers0_test_path = os.path.join(path_pers,'test_pers_001/pers_test_h0.npy')
pers1_test_path = os.path.join(path_pers,'test_pers_001/pers_test_h1.npy')
pers2_test_path = os.path.join(path_pers,'test_pers_001/pers_test_h2.npy')
np.save(pers0_test_path,test_pers0)
np.save(pers1_test_path,test_pers1)
np.save(pers2_test_path,test_pers2)

#val
pickle_val_path = os.path.join(pickle_path,'val')
files_val = os.listdir(pickle_val_path)
#files_train.sort(key=lambda x:int(x[9:].split('.')[0]))
files_val.sort(key=lambda x:int(x[4:].split('.')[0]))
# print(len(files_train))
# print(files_train)
val_pers0 = []
val_pers1 = []
val_pers2 = []
for f1 in tqdm(files_val):
    f1_path = os.path.join(pickle_val_path,f1)
    with open(f1_path, "rb") as file:
        loaded_s = pickle.load(file)
        #load_train.append(loaded_s)
    val_pers0.append(loaded_s.representation(degrees=[0], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    val_pers1.append(loaded_s.representation(degrees=[1], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
    val_pers2.append(loaded_s.representation(degrees=[2], bandwidth=.01,plot=False, p=2, kernel="gaussian", signed=False))
val_pers0 = np.array(val_pers0)
val_pers1 = np.array(val_pers1)
val_pers2 = np.array(val_pers2)

pers0_val_path = os.path.join(path_pers,'val_pers_001/pers_val_h0.npy')
pers1_val_path = os.path.join(path_pers,'val_pers_001/pers_val_h1.npy')
pers2_val_path = os.path.join(path_pers,'val_pers_001/pers_val_h2.npy')
np.save(pers0_val_path,val_pers0)
np.save(pers1_val_path,val_pers1)
np.save(pers2_val_path,val_pers2)
```
